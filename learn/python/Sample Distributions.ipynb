```python
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt

from matplotlib import ticker

def plot(fn, *args, **kwargs):
    sample = np.array([fn(*args, **kwargs) for _ in range(10000)])
    ax = sns.distplot(sample, kde=False, label="Data")
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: x / 10000))
    
    avg = sample.mean()
    std = sample.std()
    gauss = np.array([random.gauss(avg, std) for _ in range(10000)])
    sns.distplot(gauss, kde=False, ax=ax, label="Guass")
    
    lam = 1 / avg
    exp = np.array([random.expovariate(lam) for _ in range(10000)])
    sns.distplot(exp, kde=False, ax=ax, label="Exp")
    
    ax.legend(loc='upper right', shadow=True)
```

```python
plot(random.random)
```

```python
plot(random.gauss, .5, .125)
```

```python
plot(random.lognormvariate, .1, .5)
```

```python
plot(random.expovariate, 8)
```
